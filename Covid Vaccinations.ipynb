```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sbn
```

```python
file_path = r"C:\Users\Rahi\OneDrive\Desktop\covid_vaccination\us_state_vaccinations.csv"
df = pd.read_csv(file_path)# Load the CSV file here
```

```python
missing_values = df.isnull().sum() #Check for missing values in each column
print(missing_values)
```

```python
sumstats = df.describe()  #Provides an overview of the certain stats such as mean, median, etc.
print(sumstats)
```

```python
plt.figure(figsize=(12, 8)) #this will give us the distribution of the columns
df['total_vaccinations'].dropna().plot(kind='hist', bins=50, color='royalblue')
plt.title('The Distribution of Total Vaccinations')
plt.xlabel('Total # of Vaccinations')
plt.ylabel('Frequency of Vaccinations')
plt.show()
```

```python
df['date'] = pd.to_datetime(df['date']) #another way to show the trend of daily vaccinations every day 
daily_vacc_trend = df.groupby('date')['daily_vaccinations'].sum()
```

```python
plt.figure(figsize=(12, 8))
daily_vacc_trend.plot()
plt.title('Daily Vaccinations Over a Certain Period')
plt.xlabel('Date By Month')
plt.ylabel('Number of Vaccinations')
plt.grid(True)
plt.show()
```
